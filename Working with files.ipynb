```python
lst = []
path = "C:\\Users\\Admin\\OneDrive\\Documents\\Advanced python\\Books"
for book in os.listdir("Books"):
    path1 = os.path.join(path,book)
   
    for author in os.listdir(path1):
        path2 = os.path.join(path1,author)
        for title in os.listdir(path2):
            path3 = os.path.join(path2,title)
            #print(path3)
            #print(f"{book} - {author} - {title}")
            bk = open(path3)
            text = bk.read()
            
            lst.append(''.join(str(x) for x in text))
            
#             new_file = ""
#             for file in text:
#                 new_file += file
#             print(new_file)
print(lst)
```

```python
# convert to lowercase
lstt = [x.lower() for x in lst]
print(lstt)
```

```python
# Remove ""
lstt1 = [i.replace('"', '') for i in lstt]
print(lstt1)
```

```python
# Remove \n
lstt2 = [i.replace('\n', '') for i in lstt1]
print(lstt2)
```

```python
# join the strings
lstt3 = ' '.join(lstt2)
print(lstt3)
```

```python
# Tokenization
import nltk
from nltk.tokenize import word_tokenize
clean = word_tokenize(str(lstt3))
print(clean)
```

```python
# Remove stop words
from string import punctuation
nltk.download('stopwords')
stop = nltk.corpus.stopwords.words("english") + list(punctuation)
clean = [i for i in clean if not i in stop]
print(clean)
```

```python
# Lematizing the text
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
clean =[lemmatizer.lemmatize(i) for i in clean]
print(clean)
```

```python
from nltk.probability import FreqDist
freqdist = FreqDist(clean)
freqdist
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize = (28,10))
plt.bar(list(freqdist.keys()), freqdist.values(), color='g')
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
plt.figure(figsize = (16,10))
plt.clf()
plt.imshow(wordcloud)
plt.axis('off')
wordcloud = WordCloud(stopwords = STOPWORDS, background_color = "black", max_words = 300).generate(str(clean))
plt.show()
```
